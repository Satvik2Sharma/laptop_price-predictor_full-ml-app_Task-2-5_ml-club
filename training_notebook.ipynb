# Preprocessing Begins:

```python
# Removing the column 'laptop_ID', as it is just serial number.
df.drop(columns = ['laptop_ID'], inplace = True)
```

```python
# Removing 'GB' from the whole Ram Column and Converting it to Integer for better memory allocation and clean data.
df['Ram'] = df['Ram'].str.replace('GB', '', regex=False)
df['Ram'] = df['Ram'].astype(int)
```

```python
# Removing 'kg' from the whole Weight Column and Converting it to Integer for better memory allocation and clean data.
df['Weight'] = df['Weight'].str.replace('kg', '', regex=False)
df['Weight'] = df['Weight'].astype(float)
```

```python
# Trying to Standardise the 'ScreenResolution' Column
df["ScreenResolution"].value_counts()
```

```python
df['resolution'] = df['ScreenResolution'].str.split().str[-1]

df['X_res'] = df['resolution'].str.split('x').str[0].astype(int)
df['Y_res'] = df['resolution'].str.split('x').str[1].astype(int)
```

```python
df['Touchscreen'] = df['ScreenResolution'].apply(
    lambda x: 1 if 'Touchscreen' in x else 0
)
```

```python
df['IPS'] = df['ScreenResolution'].apply(
    lambda x: 1 if 'IPS' in x else 0
)
```

```python
def get_screen_type(x):
    if '4K' in x or 'Ultra HD' in x:
        return '4K'
    elif 'Full HD' in x:
        return 'Full HD'
    elif 'HD' in x:
        return 'HD'
    else:
        return 'Other'

df['Screen_Type'] = df['ScreenResolution'].apply(get_screen_type)
```

```python
df.drop(columns=['ScreenResolution', 'resolution'], inplace=True)
```

```python
# Using Numpy, Creating a pixel density column for better price prediction
df['ppi'] = (((df['X_res']**2 + df['Y_res']**2)**0.5) / df['Inches'])
```

```python
df.drop(columns=['X_res','Y_res','Inches'], inplace=True)
```

```python
# CPU Brand
def get_cpu_brand(x):
    if 'Intel Core i7' in x:
        return 'i7'
    elif 'Intel Core i5' in x:
        return 'i5'
    elif 'Intel Core i3' in x:
        return 'i3'
    elif 'Intel' in x:
        return 'Other Intel'
    elif 'AMD' in x:
        return 'AMD'
    else:
        return 'Other'
df['Cpu_brand'] = df['Cpu'].apply(get_cpu_brand)
# CPU Speed (GHz)
df['Cpu_speed'] = df['Cpu'].apply(lambda x: float(x.split()[-1].replace('GHz','')))

df.drop(columns=['Cpu'], inplace=True)
df = pd.get_dummies(df, columns=['Cpu_brand'], drop_first=True)
```

```python
df['Gpu_brand'] = df['Gpu'].apply(lambda x: x.split()[0])
df.drop(columns=['Gpu'], inplace=True)
df = pd.get_dummies(df, columns=['Gpu_brand'], drop_first=True)
```

```python
import re
def process_memory(mem):
    ssd = 0
    hdd = 0
    items = mem.split('+')
    for item in items:
        item = item.strip()
        # Convert TB to GB
        if 'TB' in item:
            size = float(re.findall(r'\d+\.?\d*', item)[0]) * 1024
        else:
            size = float(re.findall(r'\d+\.?\d*', item)[0])
        if 'SSD' in item:
            ssd += size
        elif 'HDD' in item:
            hdd += size
    return pd.Series([ssd, hdd])
df[['SSD','HDD']] = df['Memory'].apply(process_memory)
df.drop(columns=['Memory'], inplace=True)
```

```python
def clean_os(x):
    if 'Windows' in x:
        return 'Windows'
    elif 'Mac' in x or 'macOS' in x:
        return 'Mac'
    elif 'Linux' in x:
        return 'Linux'
    else:
        return 'Other'
df['OpSys'] = df['OpSys'].apply(clean_os)
df = pd.get_dummies(df, columns=['OpSys'], drop_first=True)
```

```python
import pickle
%pip install scikit-learn
%pip install xgboost
import sklearn
from sklearn.model_selection import train_test_split, KFold, cross_val_score, RandomizedSearchCV
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import Ridge
from xgboost import XGBRegressor
```

```python
X = df.drop('Price_euros', axis=1)
y = np.log(df['Price_euros']) # Log transformss for right-skewed pricing
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
```

```python
# Defining Models to Test
models = {
    "Ridge Regression": Ridge(alpha=10),
    "Random Forest": RandomForestRegressor(random_state=42),
    "Gradient Boosting": GradientBoostingRegressor(random_state=42),
    "XGBoost": XGBRegressor(random_state=42)
}
```

```python
# 5. Baseline Cross-Validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)

for name, model in models.items():
    cv_scores = cross_val_score(model, X_train, y_train, cv=kf, scoring='r2')
    print(f"{name}: Average R2 = {cv_scores.mean():.4f} (+/- {cv_scores.std():.4f})")
```

```python
# Hyperparameter Tuning (XGBoost)
xgb_params = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'max_depth': [3, 5, 7, 9],
    'colsample_bytree': [0.5, 0.7, 1.0],
    'subsample': [0.6, 0.8, 1.0]
}
```

```python
xgb_random = RandomizedSearchCV(
    estimator=XGBRegressor(random_state=42),
    param_distributions=xgb_params,
    n_iter=20,
    cv=3,
    verbose=1,
    random_state=42,
    n_jobs=-1 # Uses all available CPU cores
)

xgb_random.fit(X_train, y_train)
print(f"Best Parameters: {xgb_random.best_params_}")
```

```python
best_model = xgb_random.best_estimator_
y_pred_log = best_model.predict(X_test)

y_test_euros = np.exp(y_test)
y_pred_euros = np.exp(y_pred_log)

mae = mean_absolute_error(y_test_euros, y_pred_euros)
rmse = np.sqrt(mean_squared_error(y_test_euros, y_pred_euros))
r2 = r2_score(y_test_euros, y_pred_euros)

print(f"R-Squared ($R^2$): {r2:.4f} (Closer to 1 is better)")
print(f"Mean Absolute Error (MAE): â¬{mae:.2f}")
print(f"Root Mean Squared Error (RMSE): â¬{rmse:.2f}")
```
